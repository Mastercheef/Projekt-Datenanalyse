Example notebook to show how the built system works on real stock data. The Stock for this example are the so-called FAANG stocks which consists of:
Facebook, Amazon, Apple, Netflix and Google

```python
faang = ['META', 'AMZN', 'AAPL', 'NFLX', 'GOOG']
labels = faang
```

Then the stocks will be used to build a stock DataFrame between a specific interval.

```python
faang_stock = Stocks(faang, labels, start="2014-01-01", stop="2021-10-01")
```

### Plot the detected anomalies in the FAANG stocks.

A plot with all anomalies found and a plot with corrected anomalies are shown in each case.
A corrected anomaly is one where at least two different features have occurred

Plotting of the wanted stocks with the detected anomalies. The isolation forest was used with the return, difference and realized semi variance feature.

```python
for label in labels:
    stockDf = pd.DataFrame(faang_stock.df_stocks[label]['Close'])
    df = build_stock(stockDf,N=1,contamin=0.008)
    plotter(df,label,False)
    plot_signed_anomaly(df,label)
```
